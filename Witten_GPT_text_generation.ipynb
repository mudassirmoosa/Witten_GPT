# Introduction:

In this notebook, our goal is to test how the trained model performs at generating text. We will use the saved model weights from each epoch so that we can analyze how the quality of the generated text changes with additional training. We will use the following two input prompts:

Input prompt 1: "We re-examine the black hole solutions in classical theories of dilaton gravity in two dimensions."

Input prompt 2: "\section{Gauge-gravity duality}"

The first prompt is copied from the test dataset.

```python
from huggingface_hub import HfApi, login, hf_hub_download
from google.colab import userdata
import os
from transformers import GPT2Tokenizer, GPT2LMHeadModel
from transformers import set_seed, pipeline
import torch

# To access the HF repo for weights
hf_token = userdata.get('witten_gpt_TOKEN')
```

```python
# Loading the tokenizer and the (BASE) model
# Your existing model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Function to generate the text from the model
# Given the input string
def generate_text(text_generator,
                  input_text,
                  weight_files_path = None,
                  temperature = 0.0,
                  num_generation = 1,
                  max_new_tokens = 256,
                  ):

  output = text_generator(input_text,
                  max_new_tokens=max_new_tokens,
                  num_return_sequences=num_generation,
                  temperature=temperature,
                  do_sample=True,
                  pad_token_id=tokenizer.eos_token_id)

  return [output[k]['generated_text'] for k in range(num_generation)]


# Function to load the model
# weights from HF
def get_weights(weight_files_path, hf_token):

  login(token=hf_token)
  # Downloading the model weights file
  repo_id = "mudassirmoosa/witten_gpt_private"
  model_file = hf_hub_download(
                  repo_id=repo_id,
                  filename=weight_files_path,
                  token=hf_token
                  )

  state_dict = torch.load(model_file, map_location='cpu')
  return state_dict


# Function to print out generated text
# with various temperatures
# for a given input and text_generator
def print_results(input_text):

  temperatures = [0.7,0.5,0.3]
  outputs = []
  for temperature in temperatures:
    output = generate_text(text_generator,input_text, temperature = temperature, num_generation = 3)
    outputs.append(output)

  print(input_text)
  for k in range(3):
    print(f'\n\nOutput with temperature {temperatures[k]}: \n\n')
    for output in outputs[k]:
      print(output,"\n\n")
```

```python
text_input_1 = "We re-examine the black hole solutions in classical theories of dilaton gravity in two dimensions. "

text_input_2 = "\section{Gauge-gravity duality}"
```

# Base model (Benchmark)

```python
# Building a pipeline
model.to(device)
text_generator = pipeline('text-generation', model=model, tokenizer=tokenizer)

print_results(text_input_1)
```

```python
print_results(text_input_2)
```
